#ðŸ”§ TransformaciÃ³n

```python
import numpy as np
```

```python

datos.info()
```

```python
datos.dtypes
```

```python
customer=pd.json_normalize(datos["customer"])
customer.head()
phone=pd.json_normalize(datos["phone"])
phone.head()
internet=pd.json_normalize(datos["internet"])
internet.head()
account=pd.json_normalize(datos["account"])
account.head()
```

```python
datos_combinados = pd.concat([datos[['customerID', 'Churn']], customer, phone, internet, account], axis=1)

display(datos_combinados.head())
```

```python
datos_combinados.dtypes
```

```python
datos_combinados["Cuentas_Diarias"]=datos_combinados["Charges.Monthly"]/30
datos_combinados.head()
```

#ðŸ“Š Carga y anÃ¡lisis

```python
datos_combinados.describe()
```

```python
import matplotlib.pyplot as plt

churn_counts = datos_combinados['Churn'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(churn_counts, labels=churn_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('DistribuciÃ³n de Churn')
plt.show()
```

```python
categorical_cols = []
numerical_cols = []

for col in datos_combinados.columns:
    if col not in ['customerID', 'Churn']:
        if datos_combinados[col].dtype == 'object':
            categorical_cols.append(col)
        elif datos_combinados[col].dtype in ['int64', 'float64']:
            numerical_cols.append(col)

print("Categorical columns:", categorical_cols)
print("Numerical columns:", numerical_cols)
```

```python
import seaborn as sns

for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=datos_combinados, x=col, hue='Churn')
    plt.title(f'Distribution of {col} by Churn')
    plt.show()

for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=datos_combinados, x='Churn', y=col)
    plt.title(f'Distribution of {col} by Churn')
    plt.show()
```
